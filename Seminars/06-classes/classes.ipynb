### `Объектно-ориентированное программирование`

- Программа, как и окружающий мир, состоит из сущностей<br><br>
- Сущности имеют какое-то внутренее состояние<br><br>
- Также сущности взаимодействуют друг с другом<br><br>
- ООП нужно для описания программы в виде сущностей и их взаимоотношений<br><br>
- При этом и на сущности, и на отношения накладываются ограничения<br><br>
- Это позволяет писать более короткий, простой и переиспользуемый код

#### `Принципы ООП`

- **Абстракция** - выделение важных свойств объекта и игнорирование прочих<br><br>

- **Инкапсуляция** - хранение данных и методов работы с ними внутри одного класса с доступом к данным только через методы<br><br>

- **Наследование** - возможность создания наследников, получающих все свойства родителей с возможностью их переопределения и расширения<br><br>

- **Полиморфизм** - возможность использования объектов разных типов с общим интерфейсом без информации об их внутреннем устройстве

#### `Базовые понятия: класс и объект`


- __Класс__ представляет собой тип данных (как int или str)<br><br>
- Это способ описания некоторой сущности, её состояния и возможного поведения<br><br>
- Поведение при этом зависит от состояния и может его изменять<br><br>
- __Объект__ - это конретный представитель класса (как переменная этого типа)<br><br>
- У объекта своё состояние, изменяемое поведением<br><br>
- Поведение полностью определяется правилами, описанными в классе<br><br>
- Класс описывается __методами__ и __атрибутами__<br><br>

### `Классы в Python`

Пример того, как выглядит класс в Python:

````python
class Car:
    def __init__(self, vin_number):
        self.tank = 0
        self.max_tank = 100
        self.is_turned_on = False
        self.vin_number = vin_number
        
    wheel_count = 4

    def fill_up(self, gas_volume):
        self.tank += gas_volume
        print(f"Tank is {(self.tank / self.max_tank) * 100}% full")

    def turn_on(self):
        self.is_turned_on = True
        print("The car was turned on")

    def turn_off(self):
        self.is_turned_on = False
        print("Turning off the car")
````

Важнейшие понятия, связанные с классами это вышеупомянутые `атрибуты` и `методы`.

**Атрибуты** — это переменные, которые хранят состояние объекта. 
    
**Методы** — это функции, определённые внутри класса. Они описывают поведение объекта.

#### `Атрибуты объекта и класса`

Атрибуты бывают: 

- Атрибуты *объекта*: Уникальны для каждого объекта (например, `vin_number`).

- Атрибуты *класса*: Общие для всех объектов этого класса (например, `wheel_count = 4`).

- Атрибуты в Python могут хранить **абсолютно любой объект**.

- Это возможно благодаря динамической типизации языка. 

- В Python **переменные** (и, соответственно, атрибуты) не привязаны к конкретному типу данных, они **являются лишь именами, которые ссылаются на объекты в памяти**. 

- Поэтому в атрибут можно записать ссылку на объект любого типа.

##### `Что могут содержать атрибуты`

**Простые типы данных**
- Числа: int, float, complex (например, 42, 3.14, 1+2j).

- Строки: str (например, "hello").

- Логические значения: bool (True, False).

- Специальный объект: NoneType (его единственное значение — None).

**Коллекции и последовательности**
- Это структуры данных, которые содержат другие объекты.

- Списки: list (например, [1, "a", True]).

- Кортежи: tuple (например, (1, "a", True)).

- Словари: dict (например, {'key': 'value', 'id': 123}).

- Множества: set, frozenset (например, {1, 2, 3}).

**Функции и методы**

- Функции: созданные через def или lambda.

- Методы: функции, привязанные к экземпляру класса.

**Классы и объекты**
- Экземпляры других классов: my_object = OtherClass().

- Сами классы: attribute = MyClass. Это позволяет создавать объекты этого класса позже.

**Прочие объекты**
- Модули: import sys; self.system_module = sys.

- Файловые объекты: self.file = open('data.txt', 'r').

- Генераторы, итераторы и другие сложные объекты.

##### `Как создавать атрибуты`

Для создания атрибутов в Python есть несколько способов:
- При описании класса в специальном методе `__init__()` - он вызывается при создании объекта класса (о методах речь пойдет дальше)
- Использовать специальный метод `__setattr__()` - вызывается при присваивании атрибуту какого-либо значения. 
    - Позволяет встроить в процесс присваивания свою логику. (о методах пойдёт речь дальше)
- Использовать специальную встроенную в Python функцию `settatr(object: class, name: string, value)`
- **Динамически** на объекте
- Создать атрибут **как атрибут класса** (то есть как общий для всех объектов класса)

````python
# Через __init__()

class Car:
    def __init__(self, vin_number): 
        self.tank = 0 # можно создавать без аргументов
        self.max_tank = 100
        self.is_turned_on = False   
        self.vin_number = vin_number # можно создавать с помощью аргументов __init__()

car_object = Car(12345)
print(f"Vin_number of the car is {car_object.vin_number}")
print(f"Current amount  of fuel in the car is {car_object.tank}")
print(f"Max amount of fuel in the car is {car_object.max_tank}")
````

````python
# Через __setattr__()

class Car:
    def __init__(self):
        pass
    def __setattr__(self, name, value):
        print(f"Установка атрибута '{name}' в значение '{value}'")
        self.__dict__[name] = value
    
car_object = Car()
car_object.color = "red"
````

````python
# Через функцию setattr()

class Car:
    def __init__(self):
        pass

car_object = Car()
setattr(car_object, "color", "red")
print(car_object.color)
````

````python
# Динамически на объекте

car_object.color = "red"
print(f"Color of the car is '{car_object.color}'")
````

````python
# Как общий атрибут

class Car:
    def __init__(self):
        pass
        
    wheel_count = 4

car_object = Car()
print(f"This class of cars has {car_object.wheel_count} wheels")
````

##### `Как получить доступ к атрибутам`

В Python для получения доступа к атрибутам возможны следующие способы:
- Через `.` после имени объекта класса (`car_object.vin_number`)
- Через специальный метод `__getattribute__()` - аналогично предыдущему, позволяет встроить свою логику. Вызывается при **любом** доступе к атрибуту.
    - `__getattr__()` вызывается только в крайнем случае, когда атрибут не был найден в `__dict__`
- Через специальную функцию `getattr(object: class, name: string)`
- Через вышеупомянутый словарь `__dict__`
- Через специальную функцию `vars()`
- Через функцию `dir()` - эта функция выводит вообще все имена в текущей области видимости (current scope)

````python
# Через .


class Car:
    def __init__(self): 
        self.tank = 0 

car_object = Car() 
print(car_object.tank)
````

````python
# Через __getattribute__()


class Car:
    def __init__(self): 
        self.tank = 0 

    def __getattribute__(self, name):
        print(f"Getting access to {name}")
        return object.__getattribute__(self, name)
    
    def __getattr__(self, name):
        print("__gettatr__() started its work")
        try:
            print(f"Got access to attr:{name}, its value is {self.__dict__[name]}")
            return "__gettatr__() finished its work"
        except:
            print("No such attr, creating one")
            self.__dict__[name] = "some value"
        return "__gettatr__() finished its work"


car_object = Car() 
print(car_object.fuel)
print(car_object.fuel)
print(car_object.tank)
````

````python
# Через getattr()

class Car:
    def __init__(self): 
        self.tank = 0 

car_object = Car() 

print(getattr(car_object, "tank"))
````

````python
# Через __dict__

class Car:
    def __init__(self): 
        self.tank = 0 

car_object = Car() 

print(car_object.__dict__["tank"])
````

````python
# Через vars()

class Car:
    def __init__(self): 
        self.tank = 0 

car_object = Car() 
print(vars(car_object))
print(vars(car_object)["tank"])
````

````python
# Через dir()

class Car:
    def __init__(self): 
        self.tank = 0 

car_object = Car() 
print(dir(car_object))
print("tank" if "tank" else None in dir(car_object))
````

##### `Как удалять атрибуты`

Для удаления атрибутов в Python есть такие способы:
- Использовать оператор `del` - cамый прямой и распространенный способ удаления атрибута с объекта или класса.
- Использовать специальную функцию `delattr(object: class, name: string)`
- По аналогии с созданием при удалении атрибута вызывается `__delattr__()` - позволяет использовать свою логику.


````python
# Через del

class Car:
    def __init__(self):
        self.vin_number = 12345
        
car_object = Car()
print(f"Vin_number of the car is {car_object.vin_number}")

del car_object.vin_number

try: 
    print(car_object.vin_number)
except:
    print("No such attr")
````

````python
# Через delattr


class Car:
    def __init__(self):
        self.vin_number = 12345
        
car_object = Car()

print(f"Vin_number of the car is {car_object.vin_number}")
delattr(car_object, "vin_number")
try: 
    print(car_object.vin_number)
except:
    print("No such attr")
````

````python
# Через __delattr__()

class Car:
    def __init__(self):
        self.vin_number = 12345
    def __delattr__(self, name):
        print(f"Удаление атрибута {name}")
        del self.__dict__[name]

car_object = Car()
print(f"Vin_number of the car is {car_object.vin_number}")

del car_object.vin_number

try: 
    print(car_object.vin_number)
except:
    print("No such attr")
````

##### `Неполнота инкапсуляции`

Выше было упомянуто, что инкапсуляция в Python отличается от таковой в классическом понимании (которое, например реализовано в C++)

Так как же проявляется это отличие?

Классическая инкапсуляция подразумевает наличие `private` и `protected` атрибутов, доступ к которым строго ограничен.

В Python есть свои аналоги (соглашения) для public, protected и private:
- Чтобы указать, что атрибут является `public` не нужно ничего делать
- Чтобы указать, что атрибут является `protected` нужно поставить перед ним один знак `_`
- Чтобы указать, что атрибут является `private` нужно поставить перед ним два знака `_`, т.е. `__`

Не зря выше написано, что это *лишь соглашения*. На деле подобные приписки никак не ограничивают доступ к данным атрибутам и **к ним всё ещё можно получить доступ извне.**

Однако просто получить доступ к `private` (`__`) атрибутам просто так не выйдет, они подвержены так называемому механизму искажения имён (name mangling).

- Таким образом, Python автоматически изменяет имя такого поля, чтобы избежать случайных конфликтов имен в дочерних классах.

- Имя `__private_variable` внутри класса `MyClass` будет преобразовано в `_MyClass__private_variable`.

````python
class Car:
    def __init__(self):
        self._owner_name = "Tom"
        self.__owner_address = "Willow street, house 1"

    def get_private(self):
        return self.__owner_address

car_object = Car()

print(f"The car`s owner name is: {car_object._owner_name}")

try:
    print(car_object.__owner_address)
except:
    print("Не могу получить доступ к запрашиваемому атрибуту")

# Но доступ все еще возможен через искаженное имя
print(car_object._Car__owner_address) 
````

#### `Методы экземпляра класса`

В Python существует три вида методов, которые можно определить внутри класса:

- Методы *экземпляра* (Instance Methods)
    - Обычные методы 
    - "Магические" методы (Class magic methods)

- Методы *класса* (Class Methods)

- *Статические* методы (Static Methods)

Они различаются тем, к чему они привязаны и какие данные могут изменять: данные конкретного объекта, данные всего класса или вообще никакие данные.

##### `Параметр self`

- Метод класса отличается от обычной функции только наличием объекта `self` в качестве первого аргумента <br><br>

- Аналогичный параметр в C++ это указатель `this`<br><br>

- Название `self` является общим соглашением, но можно использовать и другое (не стоит)<br><br>

- Метод класса, не получающий на вход `self` является _статическим_<br><br>

- Статические методы часто используются для специализированного создания объектов класса<br><br>

- У _методов класса_ входом является класс как объект<br><br>

##### `Методы объекта (Instance Methods)`

- Работают с конкретным объектом класса и имеют доступ к его атрибутам (как обычным, так и атрибутам класса).

- Всегда принимают ссылку на объект класса, который их вызвал - параметр `self`.

- Методы, получающие значения атрибутов называются `get-методами` (getters или же геттеры), а методы их устанавливающие `set-методами` (setters или же сеттеры)

 `Обычные методы`

 Такими методами являются абсолютно все функции вида `def function(self, ...)`, описанные внутри класса и не имеющие никаких *обёрток* (о них будет далее)

````python
class Car:
    def __init__(self):
        self.tank = 0
        self.max_tank = 100
        self.is_turned_on = False

    def fill_up_tank(self, amount_of_fuel): #setter
        print("Refueling car...")

        self.tank += amount_of_fuel

        if self.tank > self.max_tank:
            print("The tank is maxed out")
            self.tank = self.max_tank

        print("Refueling is complete")
        print(f"Tank is {(self.tank // self.max_tank) * 100}% full\n")

    def turn_on(self): #setter
        self.is_turned_on = True
        print("The car was turned on")

    def turn_off(self): #setter
        self.is_turned_on = False
        print("Turning off the car")
    

car_object = Car()

print(f"Curr fuel level: {car_object.tank}\n")

car_object.fill_up_tank(150)

print(f"Curr fuel level: {car_object.tank}\n")

print(f"Car is turned on: {car_object.is_turned_on}")

car_object.turn_on()

print(f"Car is turned on: {car_object.is_turned_on}")
````

`Магические методы`

Примеры основных "магических" методов объекта класса:

- `__new__(class,...)`: Самый первый метод, который вызывается для создания объекта. Он отвечает за выделение памяти под объект.

- `__init__(self,...)`: Вызывается после `__new__()` для инициализации уже созданного экземпляра (например, для присвоения начальных значений атрибутам).

- `__del__(self,...)`: Вызывается, когда сборщик мусора собирается уничтожить объект.

- `__getattribute__(self,...)`: Вызывается при любом доступе к атрибуту.

- `__setattr__(self,...)`: Вызывается при попытке присвоить значение атрибуту.

- `__delattr__(self,...)`: Вызывается при удалении атрибута.

Некоторые другие:

- `__repr__(self)`: Возвращает некоторое строковое представление объекта.

- `__add__(self, other)`: Определяет поведение для оператора сложения (+). 
    - `self`: Объект слева от знака + `other`: объект справа от знака +.
    - Метод должен возвращать результат сложения, обычно в виде нового объекта.

- `__mul__(self, other)`: Определяет поведение для оператора умножения (*). 
    - Может означать разное в зависимости от контекста: умножение на число, скалярное произведение векторов и т.д.
    
- `__call__(self, *args, **kwargs)`: Позволяет вызывать экземпляр класса так, как будто это функция.
    - Если вы определите `__call__` в своем классе, вы сможете писать `my_object()` вместо того, чтобы вызывать какой-то конкретный метод.

````python
class Car:
    def __init__(self):
        self.tank = 0
        self.max_tank = 100
        self.is_turned_on = False
    
    def __repr__(self):
        return f"This is an abstract car with {self.tank} fuel"
    
    def __add__(self, other):
        car_object = Car()
        car_object.tank = self.tank + other.tank
        return car_object

    def __mul__(self, coef):
        car_object = Car()
        car_object.tank = self.tank * coef
        return car_object
    
    def __call__(self):
        print("You were driven from A to B")
        self.tank -= 10

car_1 = Car()
car_2 = Car()
car_1.tank += 50
car_2.tank += 30

print(car_1)

print(car_1 + car_2)

print(car_1 * 5)

car_1()

print(car_1)
````

### `Что такое менеджер контекста`

Менеджер контекста (оператор `with`) работает с двумя магическими методами:

- `__enter__` - код, который нужно выполнить над объектом при входе в блок менеджера
- `__exit__` - код, который нужно в любом случае выполнить при выходе из блока

````python
class SomeDataBaseDao:
    def __init__(self): self._db = ()

    def append(self, value): self._db.append(value)

    def __enter__(self):
        self._db = list(self._db)
        print('Set DB to read-write mode')
        return self

    def __exit__(self, exception_type, exception_val, trace):
        self._db = tuple(self._db)
        print('Set DB to read-only mode')
        return True

dao = SomeDataBaseDao()
#dao.append(1)  # AttributeError: 'tuple' object has no attribute 'append'
with dao:
    dao.append(1)
print(dao._db)
````

### `Наследование в Python`

````python
class Parent:
    def __init__(self):
        self.value = 10

    def get_value(self):
        return self.value

class Child(Parent):
    pass

class NewChild(Parent):
    def __init__(self):
        super().__init__()  # without it we will not have `value` field
        self.new_value = 20

print(Parent().get_value(), Child().get_value(), NewChild().get_value())

print(Child().__dict__)
print(NewChild().__dict__)
````

#### `Перегрузка родительских методов`

````python
class Parent:
    def __init__(self, value):
        self._value = value

    def get_value(self):
        return self._value

    def __str__(self):
        return f'Value: {self._value}'

class Child(Parent):
    def __init__(self, value):
        Parent.__init__(self, value)  # == super().__init__(value)

    def get_value(self):
        return Parent.get_value(self) * 2  # == super().get_value() * 2

print(Parent(10).get_value())
print(Child(10).get_value())
print(Child(10)._value)
print(Child(10))
````

#### `Множественное наследование`

- Даёт возможность классу получить методы и свойства сразу нескольких предков<br><br>
- Позволяет строить сложные иерархии зависимостей<br><br>
- Использовать без необходимости не нужно, поскольку архитектура кода сильно усложняется<br><br>
- В Python поддерживается без ограничений<br><br>
- У класса может быть один и более предков (`object` есть всегда)

- Методы и атрибуты ищутся в следующем порядке:<br><br>
    1. Имя ищется в объекте (т.е. в его `__dict__`)
    2. Дальше в классе объекта
    3. Дальше в предках класса<br><br>
- Для поиска по предкам используется MRO (Method resolution order)<br><br>
- У каждого класса в момент создания вычисляется этот порядок и сохраняется в атрибуте `__mro__`

````python
class A():
    def method(self): return 'A'

class B():
    def method(self): return 'B'

class C(A, B): pass

print(C.__mro__)
print(C().method())
````

#### `Вызов метода класса-родителя`

- В Python есть два способа обращения к родительским методам<br><br>
    - Через функцию `super` (использует порядок mro)
    - Напрямую по имени<br><br>

````python
class A():
    def method(self): return 'A'

class B():
    def method(self): return 'B'

class C(A, B):
    def method(self):
        return (A.method(self), B.method(self), super().method())

print(C.__mro__)
print(C().method())
````

### `Декораторы в Python`

- Функции или классы, которые позволяют добавлять дополнительную функциональность к другим функциям или методам <br>
- В Python они реализованы как *обёртки* вокруг функций или методов

````python
def simple_decorator(func):
    def wrapper():
        print("Before function call")
        func()
        print("After function call")
    return wrapper

@simple_decorator
def say_hello():
    print("Hello!")

say_hello()
````

Чтобы передать аргументы в декорируемую функцию, используется `*args` и `**kwargs`:

````python
def decorator_with_args(func):
    def wrapper(*args, **kwargs):
        print(f"Arguments were: {args}, {kwargs}")
        return func(*args, **kwargs)
    return wrapper

@decorator_with_args
def greet(name):
    print(f"Hi, {name}!")

greet("Alice")
````

Можно накладывать несколько декораторов на одну функцию. Они будут применяться сверху вниз:

```
@decorator1
@decorator2
def my_func():
    pass
```
Это эквивалентно
```
my_func = decorator1(decorator2(my_func))
```

Класс — это `callable`, так что ему ничто не мешает быть декоратором:

````python
class Timer:
    from time import time
    from sys import stderr

    def __init__(self, fun):
        self.function = fun

    def __call__(self, *args, **kwargs):
        start_time = self.time()
        result = self.function(*args, **kwargs)
        end_time = self.time()
        print(f"Duration: {end_time-start_time} seconds", file=self.stderr)
        return result


# adding a decorator to the function
@Timer
def payload(delay):
    return sorted(sum(range(i)) for i in range(delay))

print(payload(10000)[-1])
````

### `Статические методы`

- Обычные методы класса принимают на вход первым параметром вызывающий объект и работают с ним
- Статические методы являются методами класса и к объекту не привязаны
- Как следствие, их можно вызывать без объекта
- В Python статический метод определяется с помощью декоратора `@staticmethod`

````python
class A:
    def method(self):
        print('Regular method')

    @staticmethod
    def static_method():
        print('Static method')

# A.method()  # method() missing 1 required positional argument: 'self'

A().method()
A.static_method()
A().static_method()  # object also has his type methods
A.method(A())
````

### `Методы класса`

- Метод класса получает на вход объект-класса вместо объекта класса
- Используется для доступа к свойствам класса в целом (а не свойствам его экземпляров)
- В Python метод класса определяется с помощью декоратора `@classmethod`

````python
class Cls:
    attr = 10

    @classmethod
    def set_attr(cls, val):
        cls.attr = val

a = Cls()
print(a.attr)

b = Cls()
b.set_attr(20)
print(a.attr)
````

### `Сохранение объектов: модуль pickle`

Метод `pickle` в Python используется для сериализации и десериализации объектов. *Сериализация* — это процесс преобразования объекта в байтовый поток, а *десериализация* — это обратный процесс, возвращающий объект из байтового потока.

````python
class Cls:
    def __init__(self, value):
        self.__value = value

    def get_value(self):
        return self.__value
````

````python
import pickle

cls = Cls(Cls(10))

with open('cls.pkl', 'wb') as fout:
    pickle.dump(cls, fout)
````

````python
with open('cls.pkl', 'rb') as fin:
    cls_2 = pickle.load(fin)
````

````python
cls_2.get_value().get_value()
````

**Ограничения и предостережения**

- Не все объекты можно сериализовать с помощью pickle. Объекты, определённые в __main__ (например, лямбда-функции), не могут быть корректно сериализованы и десериализованы <br><br>
- Будьте осторожны при десериализации данных из ненадежных источников. Модуль pickle не безопасен от произвольного кода, поэтому есть риск выполнения вредоносного кода <br><br>

### `Аннотации типов`

- Это способ явно указать ожидаемые типы для переменных, атрибутов класса и аргументов/возвращаемых значений функций<br><br>
- С Python 3.5 аннотации типов стали официальной частью синтаксиса
<br><br>

Мотивация:
- Читаемость: Улучшают понимание того, как функция или метод должен использоваться<br><br>
- Безопасность: Инструменты статического анализа (например, mypy) могут предупредить о возможных ошибках типов<br><br>
- Документация: Автоматически создаёт чёткую спецификацию для API<br><br>

````python
def greet(name: str) -> str:
    return "Hello, " + name
````

`name: str` указывает, что функция принимает строку <BR>
`-> str` указывает, что функция возвращает строку

Можно создавать элиасы на типы данных:

````python
# type Vector = list[float] since python 3.12
Vector = list[float]

def scale(scalar: float, vector: Vector) -> Vector:
    return [scalar * num for num in vector]

# passes type checking; a list of floats qualifies as a Vector.
new_vector = scale(2.0, [1.0, -4.2, 5.4])
````

````python
from typing import List
def get_lengths(strings: List[str]) -> List[int]:
    return [len(s) for s in strings]
````

````python
from typing import Dict
def get_name_mapping(names: List[str]) -> Dict[str, int]:
    return {name: len(name) for name in names}
````

````python
from typing import Union
def stringify(num: Union[int, float]) -> str:
    return str(num)
````

В этом примере num может быть как целым числом (`int`), так и числом с плавающей точкой (`float`).

````python
from typing import Optional
def find_index(nums: List[int], target: int) -> Optional[int]:
    if target in nums:
        return nums.index(target)
    return None
````

`Optional[X]` равнозначно `X | None` (или `Union[X, None]`)

````python
from typing import Callable
def apply_function(value: int, func: Callable[[int], str]) -> str:
    return func(value)
````

Здесь `func` — это функция, которая принимает `int` и возвращает `str`

````python
from typing import Iterable
def sum_values(values: Iterable[int]) -> int:
    return sum(values)
````

В этом примере `values` — это итерируемый объект, содержащий целые числа.
